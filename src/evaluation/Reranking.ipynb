```python
# %%file ./re-ranking.py
print("# ja-en")
from itertools import groupby
import pprint

file_names = []
all_scores = []
for is_bleu, g in groupby(open('../../work/all_results_ja-en.txt'), lambda line: line.startswith('BLEU')):
    if is_bleu:
        lines = list(g)
        for l in lines:
            _, scores = l.split(' = ', 1)
            BLEU_score = float(scores.split()[0])
            all_scores.append(BLEU_score)
    else:
        lines = list(g)
        for l in lines:
            file_names.append(l.strip())

score_list = [(file, score) for file, score in zip(file_names, all_scores)]
pprint.pprint(sorted(score_list, key=lambda t:t[1], reverse=True))
```

```python
import pandas as pd
baseline_path = '../../data/human_eval/aspecrt-en-ja/8000-9999-99999999-999999/8000-9999_aspec-en-ja-adequacy.txt'
proposed_path = '../../data/human_eval/aspecrt-en-ja/8002-9999-99999999-999999/8002-9999_aspec-en-ja-adequacy.txt'
base_columns = ['src', 'ref', 'base_out', 'base_score1', 'base_score2']

df_base = pd.read_csv(baseline_path, sep='\t', names=base_columns, index_col=False)
df_base
```

```python
proposed_columns = ['src', 'ref', 'proposed_out', 'proposed_score1', 'proposed_score2']
df_proposed = pd.read_csv(proposed_path, sep='\t', names=proposed_columns, index_col=False)
df_proposed
```

```python
(df_proposed[['score1', 'score2']] == df_base[['score1', 'score2']]).value_counts()
```

```python
df_compare = pd.concat([df_base, df_proposed], axis=1)
df_compare
```

```python
# score1が違う例
df_score1_diff = df_compare[~(df_proposed['proposed_score1'] == df_base['base_score1'])][['src', 'ref', 'base_out', 'proposed_out', 'base_score1', 'proposed_score1']]
df_score1_diff
```

```python
# score2が違う例
df_score2_diff = df_compare[~(df_proposed['proposed_score2'] == df_base['base_score2'])][['ref', 'base_out', 'proposed_out', 'base_score2', 'proposed_score2']]
df_score2_diff
```

```python
df_diff = pd.concat([df_score1_diff, df_score2_diff], axis=1, join='outer')
df_diff
```

```python
df_diff.to_csv('../../data/human_eval/en-ja_diff_between_1_and_7.tsv', sep='\t')
```
